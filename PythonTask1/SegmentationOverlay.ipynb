```python
import SimpleITK as sitk
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
mri_scan = sitk.ReadImage("T2W.nii")
segmentation = sitk.ReadImage("Segmentation-Segment_1-label.nii")

def simple_sitk_show(img, seg=None, slice_idx=0, title=None):
    mri_array = sitk.GetArrayFromImage(img)
    seg_array = sitk.GetArrayFromImage(seg)
    mri_array = mri_array[slice_idx, :, :]
    seg_array = seg_array[slice_idx, :, :]

        
    overlay = np.where(seg_array != 0, mri_array, 0)
    
    
    plt.imshow(overlay, cmap='gray', interpolation='none')
    plt.show()


simple_sitk_show(mri_scan, seg=segmentation, slice_idx=9)  


```
